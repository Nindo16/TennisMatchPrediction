```python
data.describe(include=['O'])
```

Le type de certaines variables n'est pas correct, elles ont été classées comme Objet alors que ce sont des entiers ou flottants. 

```python
data.L2.unique()
```

La présence de ' ' empêche la variable d'être catégorisé comme flottant. On a le même souci sur les autres variables. On va donc imputer des valeurs manquantes à la place.

```python
data.loc[data.L3 == ' ', ['L3']] = np.nan
data.loc[data.W2 == ' ', ['W2']] = np.nan
data.loc[data.W3 == ' ', ['W3']] = np.nan
data.loc[data.L2 == ' ', ['L2']] = np.nan
data.L2.unique()
```

```python
data.Lsets.unique()
```

```python
data.loc[data.Lsets == '`1', ['Lsets']] = 1
data.loc[data.EXW == '2.,3', ['EXW']] = np.nan
data.loc[data.LRank == 'NR', ['LRank']] = np.nan
data.loc[data.WRank == 'NR', ['WRank']] = np.nan
```

On peut désormais convertir correctement toutes nos colonnes. 

```python
colonnes = ['L2', 'L3', 'LRank', 'Lsets', 'W2', 'W3','EXW','WRank']

for elements in colonnes:
    data.loc[:, [elements]] = np.float64(data[elements][:])

data.describe(include=['O'])
```

```python
data.dropna(subset=['WRank', 'LRank', 'Wsets', 'Lsets', 'L1', 'W1', 'L2', 'W2'], axis=0, inplace=True)
```

On supprime les matchs ayant des valeurs manquantes dans les colonnes n'en présentant pas beaucoup.

Nous avons à notre disposition des cotes de différents sites. Dans le but de créer un modèle simple et de réduire le nombre de colonnes, nous allons supprimer tous ces indicateurs de cote et les remplacer par l'écart-type par rapport aux différentes cotes, le minimum de cotation d'un joueur, ainsi que recalculer les valeurs des autres colonnes déjà présente, le maximum et la moyenne des différentes cotes. (On recalcule là où il y avait des valeurs manquantes.) 

```python
odds_W = ['B365W', 'EXW', 'PSW', 'LBW', 'CBW', 'SJW', 'IWW', 'UBW', 'SBW', 'GBW', 'B&WW']
odds_L = ['B365L', 'EXL', 'PSL', 'LBL', 'CBL', 'SJL', 'IWL', 'UBL', 'SBL', 'GBL', 'B&WL']

data.dropna(subset=odds_W, thresh=2, axis=0, inplace=True)
data.dropna(subset=odds_L, thresh=2, axis=0, inplace=True)

data['AvgW'].fillna(data[odds_W].mean(axis=1), inplace=True)
data['AvgL'].fillna(data[odds_L].mean(axis=1), inplace=True)

data['MaxW'].fillna(data[odds_W].max(axis=1), inplace=True)
data['MaxL'].fillna(data[odds_L].max(axis=1), inplace=True)

data['StdW'] = data[odds_W].std(axis=1)
data['StdL'] = data[odds_L].std(axis=1)

data['MinW'] = data[odds_W].min(axis=1)
data['MinL'] = data[odds_L].min(axis=1)

data.drop(odds_W, axis=1, inplace=True)
data.drop(odds_L, axis=1, inplace=True)
```

Il nous faut créer notre variable cible "gagnant"

| Winner | Loser | gagnant |
| :- | -: | :-: |
| Federer | Nadal | 0

0 : si c'est le joueur 1 qui gagne  (celui de la colonne Winner)

1 : si le joueur 2 gagne            (celui de la colonne Loser)

Nous sommes confrontées à un problème, il n'y aura que des 0 vu que le joueur 1 est toujours le Winner. On va donc procéder ainsi, créer une colonne en inversant les joueurs et changeant de ce fait le résultat sur la variable cible. 

| Winner | Loser | gagnant |
| :- | -: | :-: |
| Federer | Nadal | 0
| Nadal | Federer | 1

Il faut également prendre en compte ce changement sur les autres colonnes du dataset. En faisant cela, nous aurons une variable cible contenant des 0 et des 1, on se ramène à un problème de classification binaire. 

```python
copie  = data.copy()
copie[[ 'Winner','WRank','WPts','MaxW','AvgW','StdW','MinW','Loser','LRank','LPts','MaxL','AvgL','StdL','MinL']]\
=copie[['Loser','LRank','LPts','MaxL','AvgL','StdL','MinL','Winner','WRank','WPts','MaxW','AvgW','StdW','MinW']]
```

```python
joueur_2_gagnant = [1 for a in range(data.shape[0])]
data.insert(data.shape[1], "gagnant", joueur_2_gagnant, True)

joueur_1_gagnant = [0 for a in range(copie.shape[0])]
copie.insert(copie.shape[1], "gagnant", joueur_1_gagnant, True)

data = data.append(copie)
data = data.reset_index().drop(columns=['index'])
data
```

### Encodage des variables catégoriques pour entamer la mise en place de notre modèle

```python
def encodage(df):
    for colonnes in df.select_dtypes('object').columns: 
        dummy = pd.get_dummies (df[colonnes])
        df = pd.concat( [df, dummy ], axis = 1)
        del df[colonnes]
    return df
```

```python
data_final = encodage(data_final)
```

```python
data_final
```
